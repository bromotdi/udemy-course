```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
values = [0, 1, 2, 3, 4, 5, 6, 7]
squares = [value ** 2 for value in values]
plt.plot(values, squares)
plt.show()
```

```python
plt.scatter(values, squares)
plt.show()
```

```python
values = list(range(100))
pows = [pow(value, 3) for value in values]
plt.scatter(values, pows)
plt.show()
```

```python
# scatter из библиотеки Matplotlib для создания точечного графика
# values: массив х координат
# pows: массив у координат
# c: цвет точек установлен в pows -> цвет каждой точки будет определен соответствующим значением из массива pows. Это создает цветовую градацию на графике.
# cmap: цветовая карта - отображения скалярных значений в цвета
# edgecolor: цвет краев маркеров
# s: размер маркера

plt.scatter(values, pows, c=pows, cmap=plt.cm.Blues, edgecolor='none', s=10)
plt.show()
```

```python
values = list(range(50))
squares = [pow(value, 2) for value in values]
cubes = [pow(value, 3) for value in values]

# subplots - подграфики, 2 графика по вертикале, но 1 в столбце
# subplots - возращает последовательность, которую можно передать
# sharex - ось х общая (значения values)
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.scatter(values, squares)
ax1.set_title('Square numbers')
ax2.plot(values, cubes)
ax2.set_title('Cube numbers')
plt.show()
```

```python
values = list(range(10))
squares = [pow(value, 2) for value in values]
cubes = [pow(value, 3) for value in values]

# subplots - подграфики, 1 график по вертикале, но 2 в столбце
# shareу - ось у общая (значения squares и cubes)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.scatter(values, squares)
ax1.set_title('Square numbers')
ax2.plot(values, cubes)
ax2.set_title('Cube numbers')
plt.show()
```

### Примеры plot диаграмм - line, bar, area, pie

```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 150, 130, 180, 210, 230]
}

df = pd.DataFrame(data)
# 
df.plot(x='Year', y='Sales', kind='line')
plt.title('Yearly Sales')
plt.show()
```

```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 150, 130, 180, 210, 230]
}

df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='bar')
plt.title('Yearly Sales')
plt.show()
```

**`kind`** - тип графика:
- `'line'`: линейный график (по умолчанию).
- `'area'`: график с заполненной областью под кривой.
- `'scatter'`: точечный график.
- `'bar'` или `'barh'`: cтолбчатая диаграмма (вертикальная или горизонтальная).
- `'hist'`: гистограмма.
- `'box'`: ящик с усами (боксплот).
- `'pie'`: круговая диаграмма.



```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 150, 130, 180, 210, 230]
}

df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='area')
plt.title('Yearly Sales')
plt.show()
```

```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 150, 130, 180, 210, 230]
}

# labels=df['Year'] - именнованый аргумент - серия
# autopct='%1.1f%%' - отображения процентов

df = pd.DataFrame(data)
df.plot(y='Sales', kind='pie', labels=df['Year'], autopct='%1.1f%%')
plt.title('Yearly Sales')
plt.show()
```

```python
data = {
    'Age': [25, 28, 32, 41, 42, 42, 50, 59, 60]
}

# гистограма показывает частоту определенных значений
df = pd.DataFrame(data)
df.plot(y='Age', kind='hist')
plt.title('Age distribution')
plt.show()
```

### Пример диаграммы boxplot

```python
data = {
    'Category': ['FIRST', 'FIRST', 'SECOND', 'SECOND', 'THIRD', 'THIRD', 'SECOND'],
    'Value': [10, 20, 25, 45, 45, 50, 50]
}

df = pd.DataFrame(data)
df
```

```python
# by='Category' - cтолбец по которому будут сгруппированы данные. 
# для каждой уникальной группы в столбце 'Category' будет построен ящик с усами.
# column='Value' - столбец который содержит числовые значения, на основе которых будут построены ящики с усами.

df.boxplot(by='Category', column='Value')
plt.title('Values distribution by category')
plt.show()

# зеленная линия - среднее значение для группы
# черные усики - диапазон
```

### Пример диаграммы heatmap

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Week 1 Temp': [6, 10, 12, 14, 17],
    'Week 2 Temp': [8, 12, 13, 14, 18],
    'Week 3 Temp': [9, 11, 14, 16, 20],
    'Week 4 Temp': [10, 15, 10, 17, 22]
}

df = pd.DataFrame(data)
df.head()
```

```python
# делаем 'Month' - индексом по оси 0

df.set_index('Month', inplace=True)
df.head()
```

```python
# annot=True - числовые данные в каждом квадрате

sns.heatmap(df, cmap='coolwarm', annot=True)
plt.title('Weekly temperature data')
plt.show()
```
